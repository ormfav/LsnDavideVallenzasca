## The excercise
There exists a concrete way to visualize the fact that NN are universal fitters for 2D real functions too. Starting from the fact that two neurons in a single hidden layer are able to reproduce a one dimensional bin we can extend the reasoning to two dimension: a 2D bin becomes a stripe in 3D, so creating two bins, one for the x-coordinate and the other for the y-coordinate, produces two ortogonal tripes in the space with a bump in the area of intersection. By adding a second layer with a sigle neuron that activates only when a certain treshold is exceeded, we can preserve only the bump, that is a 3D bin. Following the same procedure we can produce an arbitrary number of bins to approximate the function.

Here we want to fit 
\begin{equation}
f(x,y) = \sin(x^2+y^2)
\end{equation}

```python
def f(x):
    x2=x[:,1]
    x1=x[:,0]
    return np.sin(x1**2+x2**2)
```

We produce the dataset

```python
np.random.seed(0)
sigma=0.5


x_train = np.random.uniform(-1.5, 1.5, (ntrain,2))
y_train = np.random.normal(f(x_train), sigma) 

x_valid = np.random.uniform(-1.5, 1.5,(ntrain // 10,2))
y_valid = np.random.normal(f(x_valid), sigma)

x_target = x_valid
y_target = f(x_target)
```

And fix the parameters for training the NN

```python
ntrain=20000
batch_size=32
epochs=60
```

We did some unsuccesfull attempts (not shown here) to fit the function. We found out that by increasing the number of layers the fit got better. The difference was bigger than that due to a change in the number of neurons: this fact seems to suggest that for multidimensional functions increasing the number of layers has a greater impact on the performances of the model (in fact, even the two-layers model worked quite poorely) than increasing the number of neurons. Noticing that the error was still decreasing, we increased the number of epochs from 30 (previous excercise) to 60.

```python
model=compile_model([25,25,25],'tanh','mse','sgd')
history = model.fit(x_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        verbose=1,
        shuffle=True,
        validation_data=(x_valid, y_valid))
```

```python
test_model(model)
show_results(model,history)
```

#### Reading and printing config files

```python
#Load and print config files
def LoadConf(path):
    d={}
    with open(path) as file:
        for line in file:
            (directive,key, val) = line.split()
            if(directive=="#define"): d[key] = int(val)
    return d

def PrintConf(d):
    for key, val in d.items(): print("{0:<19}{1}".format(key,val))
        
#Plot shortcuts
def PlotSetup():
    plt.axes().set_facecolor(bg1)
    plt.grid(alpha=0.3)

def MultiplotSetup(fig,r,c):
    for i in range(r*c):
        plt.subplot(r,c,i+1)
        fig.axes[i].set_facecolor(bg1)
        plt.grid(alpha=0.3) 
    
def Legend():
    leg = plt.legend(facecolor=bg2)
    for text in leg.get_texts():
        text.set_color(el1)
        
def Errbar1Sigma(x,y,e):
    plt.vlines(x,y-e,y+e,color=er1,alpha=0.8, label="$1\sigma$")
    plt.plot(x,y,color=el1)
    
def Errbar2Sigma(x,y,e):
    plt.vlines(x,y-e,y+e,color=er1,alpha=0.8, label="$1\sigma$")
    plt.vlines(x,y-2*e,y+2*e,color=er2,alpha=0.4, label="$2\sigma$")
    plt.plot(x,y,color=el1)
```

## Introductory notes

### Implementation notes
I've implemented an awk script in order to omogenize input data (for example, space was used as separator but some cities had spaces in their name).

Each excercise is a completion of the previous so in the following we wont distinguish between excercises: everything we will talk about will surely be found in 10.3.
Parallelizing our algorithm we had to deal with two problems: 
1. find the node with best results and tell it to print
2. implementing random migration.

In order to solve the first problem we used the MPI function  `MPI_Allreduce` together with `MPI_MINLOC`: this is a global blocking function that collect inputs from all nodes and gives in output to all nodes the minimum value and the node that owns it. With this functions we are able to compare the costs of the best indiviual of each node.

The second problem was a bit more tough. We decided to use `MPI_Scatter`, that is a global blocking function that takes in input an array from one node and gives in output the i-th element to the i-th node. We created a vector of lenght N_NODES whith elements $0\dots$N_NODES-1 (that is an array, since the underlying array of a vector can be accessed thanks to the method `data()`). We shuffled the vector: the i-th node sends to the node labelled by the i-th element of the vector. Starting from here we created another vector to tell each node who is sending data to them. We used two `MPI_Scatter` to tell each node who they are sending-to and receiving-from. Now we can use `MPI_Isend` and `MPI-Ireceive` to exchange the chromosome (that is: exchanging the underlying array). We must use non-blocking instruction in order to avoid the headaches of cooridinating each exchange. In order to avoid that a chromosome is overwritten before being sent, on arrival the new chromosome are tored in a temporary array. We set an `MPI_Barrier` to be sure that all nodes waits for the communication to end before going further. Only after everyone has sent and received their chromosome each node can store their value in the population and go on independetly till next migration.

```python
fig=plt.figure(figsize=(16, 8))
plt.suptitle('Cost circle', weight='bold')

#Circle
plt.axes().set_facecolor(bg1)
##Best cost
data = np.genfromtxt("out/101-bestcost.csv", delimiter=",", skip_header=1)
x=data[:,0]
y=data[:,1]
plt.plot(x,y,color=el1, label="Cost of the best individual")
##Ave cost
data = np.genfromtxt("out/101-besthalfcost.csv", delimiter=",", skip_header=1)
x=data[:,0]
y=data[:,1]
plt.plot(x,y,color=el2, label="Cost averaged over the best half of population")

plt.xlabel('#Generation')
plt.ylabel('Cost')
plt.title("Circle")
plt.grid(alpha=0.3)
Legend()


plt.show()

```
